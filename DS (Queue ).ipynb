```python
bb=[]
def Enter():
    xx=input("enter element")
    bb.append(xx)
    print(xx,"IS ADDED")
def remove():
    if not bb:
        print("queue is empty")
    else:
        a=bb.pop(0)
        print("remove Element",a)
def display():
    print(bb)
while True:
    print("Enter Choice press 1 for add press 2 for remove press 3 for display press 4 for exit")
    bawa=int(input())
    if bawa==1:
        Enter()
    elif bawa==2:
        remove()
    elif bawa==3:
        display()
    elif bawa==4:
        break
    else:
        print("Enter Correct Choice")
```
