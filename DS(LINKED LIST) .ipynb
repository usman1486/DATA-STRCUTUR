# Linked List 
# Dynamic Data Strcuture 


```python
class Node:
    def __init__(self,data):
        self.data=data
        self.ref=None
class LinkedList:
    def __init__(self):
        self.head=None
    def print_list(self):
        if self.head is None:
            print("Linked List is empty")
        else:
            n=self.head
            while n is not None:
                print(n.data)
                n=n.ref
    def add_begin(self,data):
        new_node=Node(data)
        new_node.ref=self.head
        self.head=new_node
```

```python
bc=LinkedList()
```

```python
bc.add_begin(20)
bc.add_begin(30)
bc.add_begin(40)
```
